```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'])
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'])
```

```python
df['DATA MES'] = df['DATA FINAL'].apply(lambda x : '{}'.format(x.year)) + df['DATA FINAL'].apply(lambda x : '-{:02d}'.format(x.month))
```
